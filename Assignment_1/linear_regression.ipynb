```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
raw = pd.read_csv("01-sea-temperature.csv", header=0)
raw.head()
```

we can solve the following system of equation with linear algebra:
$$y_i = b + mx_i+e_i $$

By deffining tha matices:



$$ Y = \begin{bmatrix} y_1 \\ y_2 \\ \vdots \\ y_n \end{bmatrix}, X = \begin{bmatrix} 1 & x_1 \\ 1 & x_2 \\ \vdots & \vdots \\1 & x_n \end{bmatrix}, A = \begin{bmatrix} b \\ m \end{bmatrix}, E = \begin{bmatrix} e_1 \\ e_2 \\ \vdots \\ e_n \end{bmatrix} $$

We get the following matrix equation to solve for A:

$$Y = XA+E$$

This equation can be solve by:

$$A = (X^TX)^{-1}X^TY$$



```python
"""Generic Dataset class"""
class Dataset:
    def __init__(self, rawX: np.array , rawY: np.array):
        self.rawX = rawX
        self.rawY = rawY
    def plot_data(self):
        plt.plot(self.rawX, self.rawY, 'o')

```

```python
""""Liniar regression class"""
class Lin_reg_model(Dataset):
    def __init__(self, rawX, rawY):
        super().__init__(rawX,rawY)
        self.Y = self.rawY.reshape((-1,1))
        self.X = np.asanyarray([np.ones(len(self.rawX)), self.rawX]).T
        #concatenate a colums of ones, and                                                                                                       transpose it
    def train(self):
        U = self.X.T@self.X #intermidate step
        self.A = np.linalg.inv(U)@self.X.T@self.Y  #soves fo r A
        self.b = self.A[0]
        self.m = self.A[1]

    def print_fun(self):
        print('The regretion equietion is: \n \n y = '+ str(self.b) + ' + ' + str(self.m)+ 'x')

    def plot_model(self):
        plt.plot(self.rawX,self.b+self.m*self.rawX)


```

```python
salinity = raw.salinity.to_numpy()
temperature = raw.temperature.to_numpy()
model = Lin_reg_model(salinity, temperature)
model.train()
model.print_fun()
```

```python
"""Plot data and final equation"""
model.plot_data()
model.plot_model()
plt.title('Saliity vs Temperature')
plt.xlabel('Salinity')
plt.ylabel('Temerature')
```
